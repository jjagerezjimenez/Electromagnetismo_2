```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Función para calcular el patrón de radiación en el plano XY con rotación
def radiation_pattern_xy(angle_of_rotation_deg):
    # Convertir ángulo de rotación a radianes
    alpha = np.radians(angle_of_rotation_deg)
    
    # Ángulo phi en el plano XY (de 0 a 2*pi)
    phi = np.linspace(0, 2 * np.pi, 360)
    
    # Expresión del campo eléctrico en el plano XY
    E_xy = np.cos(phi - alpha) * (1 + np.exp(1j * np.pi * np.cos(phi - alpha)) + 
                                  np.exp(1j * 2 * np.pi * np.cos(phi - alpha)) +
                                  np.exp(1j * 3 * np.pi * np.cos(phi - alpha)))
    
    # Magnitud del campo
    E_xy_magnitude = np.abs(E_xy)
    
    # Convertir a dB para el gráfico
    E_xy_dB = 20 * np.log10(E_xy_magnitude / np.max(E_xy_magnitude))
    
    return phi, E_xy_dB
```

```python
# Función para graficar el patrón de radiación en el plano XY
def plot_radiation_pattern_xy(angle_of_rotation_deg):
    phi, E_xy_dB = radiation_pattern_xy(angle_of_rotation_deg)
    
    # Graficar el patrón de radiación en coordenadas polares
    plt.figure(figsize=(6, 6))
    ax = plt.subplot(111, projection='polar')
    ax.plot(phi, E_xy_dB)
    ax.set_title(f'Patrón de Radiación XY\nRotación: {angle_of_rotation_deg}°', va='bottom')
    ax.set_theta_zero_location('N')  # 0° en el norte
    ax.set_theta_direction(-1)  # Sentido horario
    ax.set_rlim(-40, 0)  # Limitar el rango de dB
    plt.show()
```

```python
# Función para calcular la directividad D
def calculate_directivity(angle_of_rotation_deg):
    # Integrar el patrón de radiación en 3D
    alpha = np.radians(angle_of_rotation_deg)
    theta = np.linspace(0, np.pi, 180)
    phi = np.linspace(0, 2 * np.pi, 360)
    
    # Calcular el patrón de radiación en función de theta y phi
    E_pattern = np.zeros((len(theta), len(phi)))
    
    for i, t in enumerate(theta):
        for j, p in enumerate(phi):
            E_pattern[i, j] = np.abs(np.cos(t - alpha) * (1 + np.exp(1j * np.pi * np.cos(t - alpha)) +
                                                          np.exp(1j * 2 * np.pi * np.cos(t - alpha)) +
                                                          np.exp(1j * 3 * np.pi * np.cos(t - alpha))))
    
    # Normalizar el patrón de radiación
    E_max = np.max(E_pattern)
    E_pattern_normalized = E_pattern / E_max
    
    # Calcular la directividad
    d_theta = np.pi / len(theta)
    d_phi = 2 * np.pi / len(phi)
    integral = np.sum(E_pattern_normalized**2 * np.sin(theta[:, None])) * d_theta * d_phi
    
    D = 4 * np.pi / integral
    return D
```

```python
# Función para calcular la ganancia G
def calculate_gain(angle_of_rotation_deg, efficiency=1.0):
    D = calculate_directivity(angle_of_rotation_deg)
    G = efficiency * D
    return G
```

```python
# Ejemplo de cálculo de directividad y ganancia con rotación de 20 grados
directivity = calculate_directivity(20)
gain = calculate_gain(20, efficiency=0.9)  # Suponiendo una eficiencia del 90%

print(f"Directividad (D) = {directivity:.2f}")
print(f"Ganancia (G) = {gain:.2f}")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Parámetros de la antena
lambda_ = 1  # Longitud de onda
d = lambda_ / 2  # Separación entre dipolos
k = 2 * np.pi / lambda_  # Número de onda
alpha = np.deg2rad(20)  # Ángulo de rotación en radianes

# Definir los ángulos theta y phi
theta = np.linspace(0, 2 * np.pi, 360)  # Desde 0 hasta 2pi para un círculo completo
phi = np.linspace(0, 2 * np.pi, 360)
Theta, Phi = np.meshgrid(theta, phi)

# Campo eléctrico en los diferentes planos

# En el plano XY
def Exy(theta, alpha):
    return np.sin(theta) * np.cos(k * d * np.cos(theta - alpha))

# En el plano ZY
def Ezy(theta, alpha):
    return np.cos(theta) * np.sin(k * d * np.sin(theta - alpha))

# En el plano ZX
def Ezx(theta, alpha):
    return np.sin(theta + alpha) * np.cos(k * d)

# Crear las gráficas de los tres planos

fig = plt.figure(figsize=(15, 5))

# Gráfico en el plano XY
ax1 = fig.add_subplot(131, polar=True)
E_xy = Exy(Theta, alpha)
ax1.contourf(Phi, Theta, np.abs(E_xy), cmap='viridis')
ax1.set_title('Plano XY')

# Gráfico en el plano ZY
ax2 = fig.add_subplot(132, polar=True)
E_zy = Ezy(Theta, alpha)
ax2.contourf(Phi, Theta, np.abs(E_zy), cmap='plasma')
ax2.set_title('Plano ZY')

# Gráfico en el plano ZX
ax3 = fig.add_subplot(133, polar=True)
E_zx = Ezx(Theta, alpha)
ax3.contourf(Phi, Theta, np.abs(E_zx), cmap='inferno')
ax3.set_title('Plano ZX')

plt.tight_layout()
plt.show()
```
